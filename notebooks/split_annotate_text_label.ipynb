```python
import pandas as pd
from sklearn.model_selection import train_test_split
import os
import shutil
```

# Set Up Data

```python
text_annot_0_200 = pd.read_csv('/content/drive/MyDrive/Yolov8-LicensePlate/bg udh bg/cropped_annot/text_annot_0_200.csv', index_col = 0)
text_annot_201_400 = pd.read_csv('/content/drive/MyDrive/Yolov8-LicensePlate/bg udh bg/cropped_annot/text_annot_201_400.csv', index_col = 0)
text_annot_401_600 = pd.read_csv('/content/drive/MyDrive/Yolov8-LicensePlate/bg udh bg/cropped_annot/text_annot_401_600.csv', index_col = 0)
text_annot_601_800 = pd.read_csv('/content/drive/MyDrive/Yolov8-LicensePlate/bg udh bg/cropped_annot/text_annot_601_800.csv', index_col = 0)
```

```python
print(text_annot_0_200.shape)
print(text_annot_201_400.shape)
print(text_annot_401_600.shape)
print(text_annot_601_800.shape)
```

```python
# List of dataframes to be appended
dataframes = [text_annot_0_200, text_annot_201_400, text_annot_401_600, text_annot_601_800]

# Concatenate the dataframes
annot = pd.concat(dataframes, ignore_index=True)

# Print the merged dataframe
print(annot)
```

```python
annot_train, annot_test = train_test_split(annot, test_size = 0.05, random_state=56)
```

## copy the train image to directory

```python
# Directory
destination_directory = "/content/drive/MyDrive/Yolov8-LicensePlate/PaddleOCR/crnn_train"
base_path = '/content/drive/MyDrive/Yolov8-LicensePlate/bg udh bg/cropped_train_set'

# Copy files to the destination directory
for index, row in annot_train.iterrows():
    source_filepath = os.path.join(base_path, row['Filename'])
    destination_filepath = os.path.join(destination_directory, row['Filename'])

    # Copy the file to the destination directory
    shutil.copyfile(source_filepath, destination_filepath)
```

```python
# Get the list of files in the folder
file_list = os.listdir('/content/drive/MyDrive/Yolov8-LicensePlate/PaddleOCR/crnn_train')

# Count the number of files
total_files = len(file_list)

print("Total files in the folder:", total_files)
```
