```python
import replicate
import os

image_dir = "/content/drive/MyDrive/Yolov8-LicensePlate/bg udh bg/all_img_train/Data 780"
output_dir = "/content/drive/MyDrive/Yolov8-LicensePlate/bg udh bg/all_img_train/Enhance - 2"

# Iterate over the image files in the directory
for filename in os.listdir(image_dir):
    if filename.endswith(".png"):  # Assuming all the image files have the .png extension
        image_path = os.path.join(image_dir, filename)

        output = replicate.run(
            "nightmareai/real-esrgan:42fed1c4974146d4d2414e2be2c5277c7fcf05fcc3a73abf41610695738c1d7b",
            input={"image": open(image_path, "rb")}
        )

        output_url = output

        # Construct the path to save the enhanced image in the output directory
        output_filename = os.path.join(output_dir, filename)

        # Use wget to download the image and save it with the original file name
        os.system(f'wget -O "{output_filename}" "{output_url}"')
```
