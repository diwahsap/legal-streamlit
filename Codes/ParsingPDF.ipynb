```python
import pytesseract
from PIL import Image
from pdf2image import convert_from_path
import re
import os
```

```python
def document_parsing(filepath):
  filename = os.path.basename(filepath).split('.pdf')[0]
  pages = convert_from_path(filepath)
  text_doc = ''
  for page in pages:
      text = pytesseract.image_to_string(page)
      text_doc += text

  topic_text = ''

  nomor_match = re.search(r'NOMOR\s+(\S+)', text_doc, re.IGNORECASE)
  if nomor_match:
      nomor_text = nomor_match.group(1)
  else:
      nomor_match = re.search(r"(KEPUTUSAN PRESIDEN REPUBLIK INDONESIA)\s*(NO\.[A-Z\d]+\s*TAHUN\s*\d+)", text_doc, re.IGNORECASE)
      if nomor_match:
          topic_text = nomor_match.group(1)
          nomor_text = nomor_match.group(2)
      else:
          topic_text = ""
          nomor_text = ""

  if topic_text == "":
    topic_match = re.findall(r'TENTANG\n\n(.*?)(?=\n\n)', text_doc, re.DOTALL)
    if topic_match:
        topic_text = " ".join(topic_match)
    else:
        topic_text = ""

  pasal_match = re.search(r'MEMUTUSKAN(.*?)ditetapkan di', text_doc, re.IGNORECASE | re.DOTALL)
  if pasal_match:
      pasal_text = pasal_match.group(1)
  else:
      pasal_text = ""

  if topic_text == "":
    topic_text = filename

  doc_dict = {
    "nomor": nomor_text,
    "judul": topic_text,
    "isi pasal": pasal_text
  }

  return doc_dict
```
